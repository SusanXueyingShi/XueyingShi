23) Who had more than the league average HRs in 2012 (filter out all players with less 500 PAs)? (Dataframe)

```python
yr_12_df=hit_df[hit_df['yearID']==2012]
```

```python
mean_array=yr_12_df.groupby(['lgID'])['HR'].mean()
```

```python
mean_array
```

```python
AL_df=yr_12_df[yr_12_df['lgID']=='AL']
```

```python
al_array=AL_df.groupby(['playerID']).filter(lambda x:x.HR.sum()> mean_array.iloc[0])
#[['playerID']].iloc[:,0]
                                                                                
```

```python
al_array
```

```python
NL_df=yr_12_df[yr_12_df['lgID']=='NL']
```

```python
nl_array=NL_df.groupby(['playerID']).filter(lambda x:x.HR.sum()> mean_array.iloc[1])
#[['playerID']].iloc[:,0]

```

```python
nl_array
```

```python
result_df=pd.concat([al_array, nl_array])
```
